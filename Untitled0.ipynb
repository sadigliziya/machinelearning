```python
def visualize_sample_images(dataset, class_names, num_images=5):
    """Visualize a few sample images from the dataset."""
    indices = np.random.choice(len(dataset), num_images, replace=False)  # Randomly select indices
    plt.figure(figsize=(12, 5))

    for i, idx in enumerate(indices):
        image, label = dataset[idx]
        image = image.permute(1, 2, 0)  # Change shape to (H, W, C) for displaying

        plt.subplot(1, num_images, i + 1)
        plt.imshow(image.cpu().numpy())

        # Convert the label to a standard Python integer
        class_index = int(label)  # Ensure it's an integer
        plt.title(class_names[class_index])  # Display the class name
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# Visualize some sample images from the training dataset
visualize_sample_images(training_data, class_names, num_images=5)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
def train_model(model, train_loader, test_loader, criterion, optimizer, epochs=20):
    model.train()
    train_losses, val_losses, train_accuracies, val_accuracies = [], [], [], []

    for epoch in range(epochs):
        running_loss = 0.0
        correct, total = 0, 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        train_loss = running_loss / len(train_loader)
        train_acc = 100 * correct / total
        train_losses.append(train_loss)
        train_accuracies.append(train_acc)

        val_loss, val_acc = evaluate_model(model, test_loader, criterion)[:2]
        val_losses.append(val_loss)
        val_accuracies.append(val_acc)

        print(f'Epoch [{epoch+1}/{epochs}], Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%, Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%')

    return train_losses, val_losses, train_accuracies, val_accuracies
```

```python
def plot_training_curves(train_losses, val_losses, train_accuracies, val_accuracies):
    epochs = range(1, len(train_losses) + 1)

    plt.figure(figsize=(12, 5))

    # Loss Plot
    plt.subplot(1, 2, 1)
    plt.plot(epochs, train_losses, label='Train Loss', marker='o')
    plt.plot(epochs, val_losses, label='Validation Loss', marker='o')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Loss Over Epochs')
    plt.xticks(epochs)
    plt.legend()

    # Accuracy Plot
    plt.subplot(1, 2, 2)
    plt.plot(epochs, train_accuracies, label='Train Accuracy', marker='o')
    plt.plot(epochs, val_accuracies, label='Validation Accuracy', marker='o')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy (%)')
    plt.title('Accuracy Over Epochs')
    plt.xticks(epochs)
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
def evaluate_model(model, test_loader, criterion):
    model.eval()
    running_loss = 0.0
    correct, total = 0, 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            running_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    val_loss = running_loss / len(test_loader)
    val_acc = 100 * correct / total
    return val_loss, val_acc, all_labels, all_preds
```

```python
def train_with_multiple_learning_rates(learning_rates, epochs=20):
    results = {}

    for lr in learning_rates:
        print(f"Training with Learning Rate: {lr}")

        # Data loaders
        train_loader = DataLoader(training_data, batch_size=32, shuffle=True)
        test_loader = DataLoader(test_data, batch_size=32, shuffle=False)

        # Initialize model, criterion, and optimizer
        squeezenet = models.squeezenet1_1(pretrained=True)
        squeezenet.classifier[1] = nn.Conv2d(512, 43, kernel_size=(1, 1), stride=(1, 1))
        squeezenet.num_classes = 43
        squeezenet = squeezenet.to(device)

        optimizer = optim.Adam(squeezenet.parameters(), lr=lr)
        criterion = nn.CrossEntropyLoss()

        # Train the model
        train_losses, val_losses, train_accuracies, val_accuracies = train_model(
            squeezenet, train_loader, test_loader, criterion, optimizer, epochs=epochs
        )

        # Evaluate the model on the test set
        val_loss, val_acc, all_labels, all_preds = evaluate_model(squeezenet, test_loader, criterion)

        # Store results
        results[lr] = {
            "train_losses": train_losses,
            "val_losses": val_losses,
            "train_accuracies": train_accuracies,
            "val_accuracies": val_accuracies,
            "all_labels": all_labels,
            "all_preds": all_preds
        }

        print(f'Final Test Loss: {val_loss:.4f}, Final Test Accuracy: {val_acc:.2f}%')

    return results
```

```python
def plot_confusion_matrix(labels, preds, class_names):
    cm = confusion_matrix(labels, preds)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names.values(), yticklabels=class_names.values())
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()
```

```python
def visualize_predictions(model, test_loader, num_images=5):
    model.eval()
    images, labels = next(iter(test_loader))
    images, labels = images.to(device), labels.to(device)
    outputs = model(images)
    _, preds = torch.max(outputs, 1)

    plt.figure(figsize=(15, 6))  # Adjust figure size if needed
    for i in range(num_images):  # Visualize specified number of images
        plt.subplot(1, num_images, i + 1)
        plt.imshow(images[i].permute(1, 2, 0).cpu())  # Convert image from tensor to plotable format
        pred_class = class_names[preds[i].item()]  # Get predicted class name
        true_class = class_names[labels[i].item()]  # Get true class name
        plt.title(f'Pred: {pred_class}\nTrue: {true_class}', fontsize=10)  # Use multiline titles for better spacing
        plt.axis('off')

    plt.tight_layout(pad=2)  # Adjust padding for clearer visualization
    plt.show()
```

```python
def visualize_results(results, class_names):
    for lr, result in results.items():
        print(f"Visualizing for Learning Rate: {lr}")

        # Plot confusion matrix
        plot_confusion_matrix(result["all_labels"], result["all_preds"], class_names)

        # Plot training curves
        plot_training_curves(result["train_losses"], result["val_losses"], result["train_accuracies"], result["val_accuracies"])
```

```python
learning_rates = [0.001, 0.0001]

# Train with the two different learning rates and get results
results = train_with_multiple_learning_rates(learning_rates, epochs=20)

# Visualize results
visualize_results(results, class_names)

# Visualize predictions with the best model (last trained model)
last_trained_model = models.squeezenet1_1(pretrained=True)
last_trained_model.classifier[1] = nn.Conv2d(512, 43, kernel_size=(1, 1), stride=(1, 1))
last_trained_model.num_classes = 43
last_trained_model = last_trained_model.to(device)

# Load the last trained model parameters (if saved)
# last_trained_model.load_state_dict(torch.load("best_model.pth"))
# last_trained_model.eval()

# Visualize predictions
visualize_predictions(last_trained_model, DataLoader(test_data, batch_size=32, shuffle=False))
```
